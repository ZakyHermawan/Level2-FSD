```python
import os
import random
import shutil

# Directories
od_t2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/detection-object/labels/train2017'
od_v2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/detection-object/labels/val2017'
sd_t2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/seg-drivable-10/labels/train2017'
sd_v2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/seg-drivable-10/labels/val2017'
sl_t2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/seg-lane-11/labels/train2017'
sl_v2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/seg-lane-11/labels/val2017'
img_t2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/images/train2017'
img_v2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/images/val2017'

# New directories for 20% subset
subset_img_t2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/images/train2017_20'
subset_img_v2017_dir = '/home/irman/Documents/YOLOv8-multi-task/dataset/images/val2017_20'

# Create new directories if they do not exist
os.makedirs(subset_img_t2017_dir, exist_ok=True)
os.makedirs(subset_img_v2017_dir, exist_ok=True)

# Get the list of filenames
od_t2017 = os.listdir(od_t2017_dir)
od_v2017 = os.listdir(od_v2017_dir)
sd_t2017 = os.listdir(sd_t2017_dir)
sd_v2017 = os.listdir(sd_v2017_dir)
sl_t2017 = os.listdir(sl_t2017_dir)
sl_v2017 = os.listdir(sl_v2017_dir)
img_t2017 = os.listdir(img_t2017_dir)
img_v2017 = os.listdir(img_v2017_dir)

# Function to filter labels based on images
def filter_labels(labels, images):
    image_basenames = {os.path.splitext(img)[0] for img in images}
    return [label for label in labels if os.path.splitext(label)[0] in image_basenames]

# Select 20% of images randomly
def select_images(images, percentage):
    return random.sample(images, int(len(images) * percentage))

# Select 20% of train and val images
img_t2017_20 = select_images(img_t2017, 0.2)
img_v2017_20 = select_images(img_v2017, 0.2)

# Copy selected images to new directories
for img in img_t2017_20:
    shutil.copy(os.path.join(img_t2017_dir, img), os.path.join(subset_img_t2017_dir, img))

for img in img_v2017_20:
    shutil.copy(os.path.join(img_v2017_dir, img), os.path.join(subset_img_v2017_dir, img))

# Filtered lists
od_t2017_filtered = filter_labels(od_t2017, img_t2017_20)
od_v2017_filtered = filter_labels(od_v2017, img_v2017_20)
sd_t2017_filtered = filter_labels(sd_t2017, img_t2017_20)
sd_v2017_filtered = filter_labels(sd_v2017, img_v2017_20)
sl_t2017_filtered = filter_labels(sl_t2017, img_t2017_20)
sl_v2017_filtered = filter_labels(sl_v2017, img_v2017_20)

# Write to .txt files
with open('od_t2017_filtered.txt', 'w') as f:
    f.write('\n'.join(od_t2017_filtered))

with open('od_v2017_filtered.txt', 'w') as f:
    f.write('\n'.join(od_v2017_filtered))

with open('sd_t2017_filtered.txt', 'w') as f:
    f.write('\n'.join(sd_t2017_filtered))

with open('sd_v2017_filtered.txt', 'w') as f:
    f.write('\n'.join(sd_v2017_filtered))

with open('sl_t2017_filtered.txt', 'w') as f:
    f.write('\n'.join(sl_t2017_filtered))

with open('sl_v2017_filtered.txt', 'w') as f:
    f.write('\n'.join(sl_v2017_filtered))
```
